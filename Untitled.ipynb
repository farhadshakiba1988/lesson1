```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer 
from pandas.plotting import scatter_matrix
```

```python
housing = pd.read_csv('housing.csv')
housing.head()
```

```python
train_set, train_test = train_test_split(housing, test_size=0.2, random_state=42)
train_set.shape
train_set.head()
```

```python
data = train_set.copy()
data.head()
```

```python
data.plot(kind='scatter', x='longitude',y='latitude',figsize=(10,7), alpha=0.4,
         s=data['population']/20, label='population',
          c = data['median_house_value'], cmap=plt.get_cmap('jet')
         )
```

```python
# Standard correlation coefficient [-1,1]
```

```python
corr_matrix= data.corr()
corr_matrix['median_house_value'].sort_values(ascending=False)
```

```python
features = ['median_house_value','median_income','total_rooms','housing_median_age']
scatter_matrix(data[features], figsize=(15,10))
plt.show()
```
